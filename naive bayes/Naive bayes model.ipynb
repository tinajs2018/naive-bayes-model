```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df.tail()
```

```python
#droping the data which as no influence on the survival rate of the passangers
df.drop(["Name","Parch","SibSp","Ticket","Cabin","Embarked"] ,axis='columns',inplace =True)
df.head()
```

```python
target=df['Survived']
df.drop(['Survived'],inplace=True,axis='columns')
df.head()
```

```python
target.head()
```

```python
#converting string into duumy variables
dummies=pd.get_dummies(df['Sex'])
dummies.head()
```

```python
#appending the dummy varaible into the dataframe
df[['female','male']]= dummies
df.head()
```

```python
df.tail()
```

```python
#droping the original sex column
df.drop(['Sex'],axis='columns',inplace=True)
df.head()
```

```python
#checking  if any coulm as null values
df.columns[df.isna().any()]
```

```python
df.Age
```

```python
#filing the missing values  of the age using the mean of the age data
df.Age=df.Age.fillna(df.Age.mean())
df
```

```python
#spliting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df, target ,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_test.head()
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
#you train the model
model.fit(x_train,y_train)
```

```python
#checking out the score
model.score(x_test,y_test)
```
